```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Find the corelation between Grid Status and SOC

```python
plt.plot(df['Grid status'],df['SOC'])
plt.xlabel("Grid Status")
plt.ylabel("SOC")
plt.title("Linear Relationship between Grid Status and SOC ")
```

```python
plt.scatter(df['Grid status'],df['SOC'])
plt.xlabel("Grid Status")
plt.ylabel("SOC")
plt.title("Scatter plot between Grid Status and SOC ")
```

### From the above plots we can conclude that the features Grid Status and SOC are not related to each other.We cannot conclude about the exact reationships between the two features.From the below heatmap we can see that the co-relation index between Grid Status and SOC is 0.23.
### The corelation index lies between -1 to +1 wherein -1 denotes negative corealtion and and +1 denotes positive corelation.

```python
f,ax=plt.subplots(figsize=(13,13))
corr=df.corr()
sns.heatmap(corr,annot=True,ax=ax)
```

### Find the correlations between Equivalent Cycles and SOH

```python
plt.plot(df['Equivalent cycle'],df['SOH'])
plt.xlabel("Equivalent cycle")
plt.ylabel("SOH")
plt.title("Linear Relationship between Equivalent cycle and SOH ")
```

```python
plt.scatter(df['Equivalent cycle'],df['SOH'])
plt.xlabel("Equivalent cycle")
plt.ylabel("SOH")
plt.title("Scatter plot between Equivalent cycle and SOH ")
```

### From the above plots we can conclude that the features Equivalent Cycle and SOH are related to each other such that when equivalent cycle is increased SOH is decreased.This shows the negative corelation between the two features.From the below heatmap we can see that the co-relation index between Equivalent Cycle and SOH is -0.98

### As the corelation index is close to -1 we can say that the two features are highly negatively corelated to each other.

```python
f,ax=plt.subplots(figsize=(13,13))
corr=df.corr()
sns.heatmap(corr,annot=True,ax=ax)
```
